```python
import pandas as pd

csv_path = '/Users/theoverdelhan/Documents/TRADING/MYR/global_competitor_analysis/gap_study/TSLA_5d_quotes.csv'
df_price = pd.read_csv(csv_path)

df_price['ask_price'] = df_price['ap']
df_price['bid_price'] = df_price['bp']

df_price['mid_price'] = (df_price['ask_price'] + df_price['bid_price']) / 2

df_price['price'] = df_price['mid_price']

df_price['time'] = pd.to_datetime(df_price['t'], unit='ns', utc=True)
df_price['time'] = (df_price['time'].astype('int64') // 10**9 ) * 1000

print(df_price['ax'].value_counts())
print(df_price['bx'].value_counts())

columns_to_drop = ['as','ap', 'bs', 'bp', 't', 'c', 'z']
df_price.drop(columns=columns_to_drop, inplace=True)

df_price
```

```python
import pandas as pd

# Calculer les différences entre ax et bx pour chaque lettre
letters = df_price['ax'].unique()  # Obtenir toutes les lettres uniques dans ax
difference_counts = {}

for letter in letters:
    count_ax = df_price[(df_price['ax'] == letter) & (df_price['bx'] != letter)].shape[0]
    count_bx = df_price[(df_price['bx'] == letter) & (df_price['ax'] != letter)].shape[0]
    difference_counts[letter] = count_ax + count_bx

# Afficher les résultats
for letter, count in difference_counts.items():
    print(f"Lettre {letter}: {count} différences")
```

```python
symbol = 'iTSLAUSDT'


csv_path_trades = '2025-04-04_trades.csv'
csv_path_trades3 = '2025-04-03_trades.csv'
csv_path_trades2 = '2025-04-07.csv'

df_trades = pd.read_csv(csv_path_trades)
df_trades2 = pd.read_csv(csv_path_trades2)
df_trades3 = pd.read_csv(csv_path_trades3)

trades_df = pd.concat([df_trades, df_trades2, df_trades3], ignore_index=True)
trades_df = trades_df[trades_df['symbol'] == symbol]

trades_df.sort_values(by='executedAt', ascending=True, inplace=True)


trades_df['time'] = trades_df['executedAt']
trades_df['entry_price'] = trades_df['executionPrice']
trades_df['quantity'] = trades_df['executionQuantity']
trades_df['direction'] = trades_df['tradeDirection']


columns_to_drop = ['executedAt', 'executionPrice', 'executionQuantity', 'tradeDirection', 'marketId']
trades_df.drop(columns=columns_to_drop, inplace=True)

trades_df.head()
```

```python
merged = pd.merge_asof(trades_df, df_price, on='time', direction='nearest')
merged = merged.dropna()
merged.head()
```

```python
min_time_price = df_price['time'].min()
max_time_price = df_price['time'].max()

count_in_interval = df_trades[(df_trades['executedAt'] >= min_time_price) & (df_trades['executedAt'] <= max_time_price)]
print(f"Nombre de timestamps dans l'intervalle de df_price : {len(count_in_interval)}")
```

```python
merged['spread_exec'] = ((merged['entry_price'] - merged['price'])) / ((merged['entry_price'] + merged['price']) / 2) * 100
merged.head()
```

```python
import plotly.graph_objects as go

merged['datetime'] = pd.to_datetime(merged['time'], unit='ms')

df_buy = merged[merged['direction'] == 'buy']
df_sell = merged[merged['direction'] == 'sell']

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df_buy['datetime'],
    y=df_buy['spread_exec'],
    mode='markers',
    name='Buy',
    marker=dict(color='green', size=4),
    hovertemplate='Time: %{x}<br>Spread Exec: %{y:.4f}%<extra></extra>'
))

fig.add_trace(go.Scatter(
    x=df_sell['datetime'],
    y=df_sell['spread_exec'],
    mode='markers',
    name='Sell',
    marker=dict(color='red', size=4),
    hovertemplate='Time: %{x}<br>Spread Exec: %{y:.4f}%<extra></extra>'
))

fig.update_layout(
    title='Spread à l’exécution dans le temps',
    xaxis_title='Heure (UTC)',
    yaxis_title='Spread Exec (%)',
    template='plotly_white',
    height=500,
    hovermode='x unified'
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

timestamp_to_plot = 1743689007065
window = 10 * 1000

fig = go.Figure()

# Lire les CSV et tracer les données pour toutes les combinaisons
letters = ['P', 'K', 'Q']
colors = ['orange', 'blue', 'green']
color_map = {('P', 'P'): 'orange', ('P', 'K'): 'blue', ('P', 'Q'): 'green',
             ('K', 'P'): 'purple', ('K', 'K'): 'red', ('K', 'Q'): 'brown',
             ('Q', 'P'): 'pink', ('Q', 'K'): 'cyan', ('Q', 'Q'): 'lime'}

for ax_letter in letters:
    for bx_letter in letters:
        # Lire le fichier CSV correspondant
        csv_file = f"filtered_{ax_letter}_{bx_letter}.csv"
        df = pd.read_csv(csv_file)
        
        # Filtrer les données dans la fenêtre de temps
        df_window = df[
            (df['time'] >= timestamp_to_plot - window) &
            (df['time'] <= timestamp_to_plot + window)
        ]
        
        # Tracer les mid_price pour chaque combinaison
        fig.add_trace(go.Scatter(
            x=df_window['time'],
            y=df_window['mid_price'],
            mode='lines',
            line_shape='hv',
            name=f"Mid Price ({ax_letter}, {bx_letter})",
            line=dict(color=color_map[(ax_letter, bx_letter)])
        ))
        
        # Trouver le point le plus proche de la timestamp_to_plot
        df_window['time_diff'] = df_window['time'].sub(timestamp_to_plot).abs()
        closest_idx = df_window['time_diff'].idxmin()
        closest_time = df_window.loc[closest_idx, 'time']
        closest_price = df_window.loc[closest_idx, 'mid_price']
        
        # Ajouter le point correspondant à la timestamp_to_plot
        fig.add_trace(go.Scatter(
            x=[closest_time],
            y=[closest_price],
            mode='markers',
            marker=dict(color=color_map[(ax_letter, bx_letter)], size=10),
            name=f"Point ({ax_letter}, {bx_letter})"
        ))

fig.update_layout(
    title=f"Mid Price autour de la timestamp {timestamp_to_plot}",
    xaxis_title="Time (ms)",
    yaxis_title="Price",
    showlegend=True
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

timestamp_to_plot = 1743689007065
window = 10 * 1000

# Filtrer les lignes où ax et bx contiennent uniquement P, K ou Q et où ax == bx
df_filtered = df_price[
    (df_price['ax'].isin(['P', 'K', 'Q'])) & 
    (df_price['bx'].isin(['P', 'K', 'Q'])) & 
    (df_price['ax'] == df_price['bx'])
]

# Filtrer les données dans la fenêtre de temps
df_window = df_filtered[
    (df_filtered['time'] >= timestamp_to_plot - window) &
    (df_filtered['time'] <= timestamp_to_plot + window)
]

fig = go.Figure()

# Tracer les trois mid_price différents
fig.add_trace(go.Scatter(
    x=df_window['time'],
    y=df_window['mid_price'],
    mode='lines',
    line_shape='hv',
    name='Mid Price',
    line=dict(color='orange')
))

# Identifier les trois points les plus proches de la timestamp_to_plot
df_window['time_diff'] = df_window['time'].sub(timestamp_to_plot).abs()
df_closest = df_window.nsmallest(3, 'time_diff')  # Trouver les 3 plus proches

# Ajouter des lignes pour les trois mid_price les plus proches
for i, row in df_closest.iterrows():
    fig.add_trace(go.Scatter(
        x=[row['time'], row['time'] + 1],  # Ligne horizontale en mode hv
        y=[row['mid_price'], row['mid_price']],
        mode='lines',
        line_shape='hv',
        name=f"Mid Price ({row['ax']})",
        line=dict(dash='dash', width=2)
    ))

fig.update_layout(
    title=f"Mid Price, Bid, et Ask autour de la timestamp {timestamp_to_plot}",
    xaxis_title="Time (ms)",
    yaxis_title="Price",
    showlegend=True
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# Timestamp cible
timestamp_to_plot = 1743689007065

# Fenêtre de 10 secondes avant et après
window = 2 * 1000  # en millisecondes

# Filtrage autour de la timestamp
df_window = df_price[
    (df_price['time'] >= timestamp_to_plot - window) &
    (df_price['time'] <= timestamp_to_plot + window)
].copy()  # copy pour éviter les SettingWithCopyWarning

# Calcul du prix de référence (mid_price au début de la fenêtre)
price_ref = df_window.iloc[0]['mid_price']
df_window['spread'] = df_window['mid_price'] - price_ref

# Création du graphique
fig = go.Figure()

# Trace mid_price
fig.add_trace(go.Scatter(
    x=df_window['time'],
    y=df_window['mid_price'],
    mode='lines',
    name='Mid Price'
))

# Point rouge à la timestamp cible
closest_idx = df_window['time'].sub(timestamp_to_plot).abs().idxmin()
closest_time = df_window.loc[closest_idx, 'time']
closest_price = df_window.loc[closest_idx, 'mid_price']

fig.add_trace(go.Scatter(
    x=[closest_time],
    y=[closest_price],
    mode='markers',
    marker=dict(color='red', size=10),
    name='Target Timestamp'
))

# Ajout de la courbe de spread
fig.add_trace(go.Scatter(
    x=df_window['time'],
    y=df_window['spread'],
    mode='lines',
    name='Spread (mid_price - price_ref)',
    yaxis='y2',
    line=dict(dash='dot', color='orange')
))

# Layout avec un deuxième axe Y pour le spread
fig.update_layout(
    title=f"Mid Price et Spread autour de la timestamp {timestamp_to_plot}",
    xaxis_title="Time (ms)",
    yaxis=dict(title="Mid Price"),
    yaxis2=dict(title="Spread", overlaying='y', side='right', showgrid=False),
    legend=dict(x=0, y=1),
    height=600
)

fig.show()
```
